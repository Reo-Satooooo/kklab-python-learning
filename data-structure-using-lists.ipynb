研究室勉強会(Python編・リストを使ったデータ構造)

```python
# 例題1：トランプカードをリストで生成する
symbol_list = ['C','D','S','H']
card_list = []
for symbol in symbol_list:
  for number in range(1,14):
    if number == 1:
      card = symbol +'A'
    elif number == 11:
      card = symbol +'J'
    elif number == 12:
      card = symbol +'Q'
    elif number == 13:
      card = symbol +'K'
    else:
      card = symbol + str(number)
    card_list.append(card)
print(card_list)
```

```python
# 例題2:作成したトランプをシャッフルし、カードの先頭を取り出す。
import random
random.shuffle(card_list)
drawn = card_list.pop(0)
print(drawn)
```

```python
#　課題1:手札を5枚取得するように書き換えなさい
card_draw = []
for number in range(0,5):
  drawn = card_list.pop(0)
  card_draw.append(drawn)
print(card_draw)
```

```python
# 課題２：for文を内包表記を用いて書き換える
card_draw = [card_list.pop(0) for number in range(0,5)]
print(card_draw)
```

```python
# 課題3:スライシングを使って書き換える
print(card_draw[1:3])
```

```python
# 課題4:札のチェックを行う操作を追加する
card_tmp = card_draw.pop(0)
card_design = card_tmp[0]
card_number = card_tmp[1:]
print(card_design,card_number)
print('Card {0} is {1} {2}'.format(0,card_design,card_number))
```
